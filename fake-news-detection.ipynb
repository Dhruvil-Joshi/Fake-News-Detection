import pandas as pd
import re
from nltk.corpus import stopwords # the for of in with
from nltk.stem.porter import PorterStemmer # loved loving == love
from sklearn.feature_extraction.text import TfidfVectorizer # loved = [0.0]
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```python
news_another_df = pd.read_csv("/kaggle/input/fake-real-news/news_dataset.csv")
news_another_df.shape
```

```python
news_df = pd.read_csv("/kaggle/input/textdb3/fake_or_real_news.csv")
news_df.shape
```

```python
news_df.head
news_df.columns
```

```python
news_another_df.head
news_another_df.columns
```

```python
news_df = news_df.fillna(' ')
news_c=news_another_df.fillna(' ')
```

```python
ps = PorterStemmer()
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [ps.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
news_df['content'] = news_df['text']
news_df['content'] = news_df['content'].apply(stemming)
```

```python
news_another_df['content'] = news_another_df['text']
news_another_df['content'] = news_another_df['content'].apply(stemming)
```

```python
X = news_df['text'].values
y = news_df['label'].values
```

```python
vector = TfidfVectorizer()
vector.fit(X)
X = vector.transform(X)
```

```python
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
train_y_pred = model.predict(X_train)
print("train accurracy :",accuracy_score(train_y_pred,y_train))
```

```python
test_y_pred = model.predict(X_test)
print("train accurracy :",accuracy_score(test_y_pred,y_test))
```

```python
input_data = X_test[0]
prediction = model.predict(input_data)
if prediction[0] == 'FAKE':
    print('Fake news')
else:
    print('Real news')
```
